```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
ccdata=pd.read_csv('creditcard.csv')
```

```python
ccdata.head()
```

```python
ccdata.shape
```

```python
ccdata.info()
```

```python
ccdata.isnull().sum()
```

```python
ccdata['Class'].value_counts()
```

```python
normal=ccdata[ccdata.Class==0]
fraud=ccdata[ccdata.Class==1]
```

```python
normal.shape,fraud.shape
```

```python
normal_sample=normal.sample(n=88)
```

```python
new_set=pd.concat([normal_sample,fraud],axis=0)
```

```python
new_set.shape
```

```python
X=new_set.drop('Class',axis=1)
Y=new_set['Class']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
X_train,X_test,Y_train,Y_test
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
acc=accuracy_score(Y_test,Y_pred)
acc
```
