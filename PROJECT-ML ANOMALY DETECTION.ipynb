```python
import pandas as pd
```

```python
df=pd.read_csv("Thyroid_Disease_Dataset_Unsupervised.csv",sep=';')
df
```

```python
df.drop(['Unnamed: 22','Unnamed: 23'],axis=1,inplace=True)
```

```python
df
```

```python
df.columns
```

```python
df['Outlier_label '].unique()
```

```python
import seaborn as sns
sns.countplot('Outlier_label ',data=df)
```

```python
X=df[df['Outlier_label ']=='n']
X
```

```python
X_=X.drop('Outlier_label ',axis=1)
```

```python
Df=df.drop('Outlier_label ',axis=1)
```

```python
from sklearn.svm import OneClassSVM
model=OneClassSVM(nu=0.05)
```

```python
model.fit(X_)
```

```python
Y_pred=model.predict(Df)
Y_pred
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
plt.scatter(df['T3_measured'],df['TT4_measured'],c=np.where(Y_pred==-1,'red','blue'),label='OneClassSVM')
plt.title('Outlier Detection Using OneClassSVM')
plt.xlabel('Triiodothyronine')
plt.ylabel('Tyroxine')
plt.legend()
plt.show()
```

```python
#ISOLATION FOREST CLASSIFIER
from sklearn.ensemble import IsolationForest
clf=IsolationForest(contamination=0.1,random_state=42)
Y_pred_Iso=clf.fit_predict(Df)
plt.scatter(df['T3_measured'],df['TT4_measured'],c=np.where(Y_pred_Iso==-1,'red','blue'),label='Isolation Forest')
plt.title('Outlier Detection Using Isolation Forest')
plt.xlabel('Triiodothyronine')
plt.ylabel('Tyroxine')
plt.legend()
plt.show()
```

```python
from sklearn.neighbors import LocalOutlierFactor
clf_loc=LocalOutlierFactor(n_neighbors=5,contamination=0.1)
Y_pred_lof=clf_loc.fit_predict(Df)
plt.scatter(df['T3_measured'],df['TT4_measured'],c=np.where(Y_pred_lof==-1,'red','blue'),label='Local Outlier Factor')
plt.title('Outlier Detection Using Local Outlier Factor')
plt.xlabel('Triiodothyronine')
plt.ylabel('Tyroxine')
plt.legend()
plt.show()
```

```python
#Guassian Mixtures
from sklearn.mixture import GaussianMixture
model=GaussianMixture(n_components=4,random_state=42)
model.fit(X_)
```

```python
Y_pred_Guassian=model.predict(Df)
plt.scatter(df['T3_measured'],df['TT4_measured'],c=np.where(Y_pred_Guassian<0.5,'red','blue'),label='Guassian Mixtures')
plt.title('Outlier Detection Using Guassain Mixtures')
plt.xlabel('Triiodothyronine')
plt.ylabel('Tyroxine')
plt.legend()
plt.show()
```

```python
print('Original Outliers in Dataset :',np.count_nonzero(df['Outlier_label ']=='o'))
#Comparision
print('Outliers Found By Each Algorithm :')
print("OneClassSVM :",np.count_nonzero(Y_pred==-1))
print("Isolation Forest :",np.count_nonzero(Y_pred_Iso==-1))
print("Local Outlier Factor :",np.count_nonzero(Y_pred_lof==-1))
print("Guassian Mixtures :",np.count_nonzero(Y_pred_Guassian<0.5))
```
